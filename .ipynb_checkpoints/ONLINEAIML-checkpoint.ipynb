```python
import pandas as pd
from mlxtend.preprocessing import minmax_scaling

# plotting modules
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('finaldata.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
label_encoded_df = df.copy()
for col in label_encoded_df.select_dtypes(include='O').columns:
    label_encoded_df[col]=le.fit_transform(label_encoded_df[col])
```

```python
label_encoded_df
```

```python
data_num = label_encoded_df[['UrbanPop','RuralPop','TotalPop','UrbanLA','RuralLA','TotalLA']]
data_num
```

```python
from numpy import asarray
from sklearn.preprocessing import StandardScaler
import statsmodels.api as sm
import matplotlib.pyplot as plt
standard_data = pd.DataFrame()
scaler = StandardScaler()
#data =data_num.values[:, :-1]
scaled = scaler.fit_transform(data_num)
scaled = pd.DataFrame(scaled,columns = ['UrbanPop','RuralPop','TotalPop','UrbanLA','RuralLA','TotalLA'])
```

```python
scaled_data = minmax_scaling(label_encoded_df,columns=['UrbanPop','RuralPop','TotalPop','UrbanLA','RuralLA','TotalLA'])

# plot both together to compare
fig, ax=plt.subplots(1,2)
sns.distplot(label_encoded_df, ax=ax[0])
ax[0].set_title("Original Data")
sns.distplot(scaled_data, ax=ax[1])
ax[1].set_title("Scaled data")
```

```python
scaled_data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
trans = MinMaxScaler()
data = trans.fit_transform(scaled_data)
data
```

```python
final_data = pd.DataFrame(data, columns =['UrbanPop','RuralPop','TotalPop','UrbanLA','RuralLA','TotalLA'])
final_data
```

```python
final_df_cat= label_encoded_df.drop(['UrbanPop','RuralPop','TotalPop','UrbanLA','RuralLA','TotalLA'],axis=1)
final_df_cat
```

```python
final_df = pd.concat([final_df_cat,final_data],axis=1)
final_df
```

```python
y = final_df['IncomeGrp']
final_df = final_df.drop(['IncomeGrp'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(final_df, y, test_size=0.2, random_state=42)
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
svc_model = SVC(kernel='linear', C=1.0)
svc_model.fit(X_train, y_train)
y_pred = svc_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
